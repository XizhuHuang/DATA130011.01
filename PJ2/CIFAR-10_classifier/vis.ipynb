```python
import torch
import matplotlib.pyplot as plt
from torchvision import datasets, transforms

# 注册 forward hook
activation = {}

def get_activation(name):
    def hook(model, input, output):
        activation[name] = output.detach()
    return hook

# 注册 block3 的输出
model.block3.register_forward_hook(get_activation('block3'))

# 准备测试集中的一张图片
transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,) * 3, (0.5,) * 3)
])

testset = datasets.CIFAR10(root='./data', train=False, download=True, transform=transform_test)
testloader = torch.utils.data.DataLoader(testset, batch_size=1, shuffle=False)

index = 109  
img, label = testset[index]  

# 将图片输入模型进行前向传播
img = img.unsqueeze(0).cuda()  # 添加 batch 维度并放到 GPU
_ = model(img)

# 获取 block3 的激活图
act = activation['block3'].squeeze().cpu()  # [C, H, W]
num_channels = min(32, act.size(0))  # 最多显示 32 个通道

# 可视化 block3 的特征图
fig, axes = plt.subplots(4, 8, figsize=(12, 6))
for i, ax in enumerate(axes.flat):
    if i < num_channels:
        ax.imshow(act[i], cmap='viridis')  # 使用 viridis 色图来显示
        ax.axis('off')

plt.suptitle(f"Feature Maps from block3 (ResidualSEBlock) - Image {index}")
plt.tight_layout()
plt.show()
```

```python
import os
import pickle
import numpy as np
from PIL import Image

# 路径配置
dataset_path = "data/cifar-10-batches-py/test_batch"  # 测试集路径

output_path = "./test_image.png"  # 输出文件路径

# 加载CIFAR-10测试数据
def unpickle(file):
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict

# 主流程
if __name__ == "__main__":
    # 1. 加载测试数据
    test_data = unpickle(dataset_path)
    
    index = 109
    image_data = test_data[b'data'][index]  # 原始字节数据 (3072维)
    image_label = test_data[b'labels'][index]  # 标签
    
    # 3. 转换图像格式 (CIFAR特殊格式 -> RGB)
    # - 分离RGB通道并重塑为32x32
    r = image_data[0:1024].reshape(32, 32)
    g = image_data[1024:2048].reshape(32, 32)
    b = image_data[2048:3072].reshape(32, 32)
    # - 合并通道并转换为uint8
    img_array = np.dstack((r, g, b)).astype(np.uint8)
    
    # 4. 保存图像
    img = Image.fromarray(img_array)
    img.save(output_path)
    
    print(f"图像已保存至: {output_path}")
    print(f"图像标签: {image_label} (类别索引)")
```

# SE 模块权重可视化

```python
import matplotlib.pyplot as plt
import torch

# 假设 se_weights 已经定义

# 提取 SE 权重的 hook
se_weights = []

def get_se_weights(name):
    def hook(module, input, output):
        w = module.squeeze(input[0]).detach().cpu()  # 注意力权重（大小为 C）
        se_weights.append(w.mean(dim=0))  # 取每个样本的平均
    return hook

model.block3.se.register_forward_hook(get_se_weights('se_weights'))  # 注册 hook

# 可视化测试集两张图片的 SE 权重
model.eval()

# 图像 1
img1, _ = testset[10]  # 获取第一张测试图片
model(img1.unsqueeze(0).cuda())  # 前向传播

# 可视化平均注意力
avg_weight1 = torch.stack(se_weights).mean(dim=0)
print(f"Average SE Weight for Image 10: {avg_weight1.shape}")

# 图像 2
se_weights.clear()  # 清除上一个图像的注意力权重
img2, _ = testset[109]  # 获取第二张测试图片
model(img2.unsqueeze(0).cuda())  # 前向传播

# 可视化平均注意力
avg_weight2 = torch.stack(se_weights).mean(dim=0)
print(f"Average SE Weight for Image 109: {avg_weight2.shape}")

# 可视化两个图像的 SE 权重
plt.figure(figsize=(16, 7))

# 第一个子图: 图像 10 的 SE 权重
plt.subplot(1, 2, 1)
plt.bar(range(len(avg_weight1)), avg_weight1.squeeze().numpy())
plt.title("SE Attention Weights for Image 10")
plt.xlabel("Channel Index")
plt.ylabel("Attention Weight")

# 第二个子图: 图像 109 的 SE 权重
plt.subplot(1, 2, 2)
plt.bar(range(len(avg_weight2)), avg_weight2.squeeze().numpy())
plt.title("SE Attention Weights for Image 109")
plt.xlabel("Channel Index")
plt.ylabel("Attention Weight")

plt.tight_layout()
plt.show()
```
